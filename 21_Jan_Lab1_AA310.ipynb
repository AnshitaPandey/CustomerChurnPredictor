```python
# importing the required libraries
from osgeo import gdal
# gdal.VersionInfo() # to check the version of the library
import os
import numpy as np
import matplotlib.pyplot as plt
import rasterio
from rasterio.windows import from_bounds
from pyproj import Transformer
```

```python
b2 = '/Users/parvathythankachyp/Downloads/AA310_satelliteimaging/Lab1/lab1_liss4/BAND2.tif'
```

```python
# Reading a band as an array using gdal
ds = gdal.Open(b2).ReadAsArray().astype(float)
ds.shape
```

```python
# Displaying the Image
plt.imshow(ds)
plt.colorbar()
```

```python
# Masking border zero value to nan
ds[ds==0]= np.nan
plt.imshow(ds, cmap='turbo')
plt.colorbar()
```

```python
# Image stretching
vmin = np.nanquantile(ds, 0.01)
vmax = np.nanquantile(ds, 0.99)
plt.imshow(ds, vmin=vmin, vmax=vmax, cmap = 'turbo')
plt.colorbar(label='DN values', shrink = 0.8, pad = 0.05, orientation = 'vertical')
plt.title("DN values (Stretched)")
plt.xlabel("Pixel location x")
plt.ylabel("Pixel location y")
plt.show()
```

```python
# Histogram
plt.hist(ds, linewidth = 1.2, alpha=0.7, rwidth = 0.85, range = (np.nanmin(ds), np.nanmax(ds)))
plt.title("Histogram of DN values, Band 2")
plt.xlabel("DN Values")
plt.ylabel("Occurence")
plt.show()
```
